### Funcion para lectura de la pagina

```python
def get_valores(tags):
    categ=''
    valor=''
    #datos['year']=nombre[0]
    datos['marca']=marca
    datos['modelo']=modelo
    datos['pasajeros']=pasajeros
    #descrip=nombre.split()[1:]
    #descrip=' '.join(descrip)
    datos['descripcion']=descrip
    datos['precio']=precio
    datos['preciodolar']=preciodolar
    
    for tag in tags:
      if tag.has_attr('class'):
        if tag['class'] == 'spec' or tag['class']==['spec']:
          valor=tag.text.strip()
        else:
          categ=tag.text.strip()

      if categ!='' and valor!='':
         datos[categ]=valor
         categ=''
         valor=''
```

```python
url = "https://crautos.com/autosusados/searchresults.cfm"
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.get(url)
soup=BeautifulSoup(driver.page_source,'html.parser')
df = pd.DataFrame()
```

```python
for pagina in range(10,20):
 print ("Pagina {}: ".format(pagina),'\n') 
 text = "javascript:p({})".format(pagina)
 
 driver.execute_script(text)
 time.sleep(3)
    
 soup=BeautifulSoup(driver.page_source,'html.parser')
 anuncios = soup.find_all('div', class_ ='inventory hidden-xs')

 for anuncio in anuncios:
   datos = {}
   nombre=anuncio.find('div').text.split() #Descripcion
   nombre=' '.join(nombre)
   datos['year']=nombre.split()[0]

   descrip=nombre.split()[2:]     
   pasajeros=nombre.split()[-2]
   nombre=nombre.split()[1:-3]
   dummy=nombre[1]
   minuscula=False
   for char in dummy:
      if char.islower():
          minuscula=True
          break
   if minuscula:
      marca=nombre[0:2]
      marca=' '.join(marca)
      modelo=' '.join(nombre[2:])
   else:
      marca=nombre[0]
      modelo=' '.join(nombre[1:])

     
   precio= anuncio.find('div',class_='pricetext').text
   precio = precio.replace('¢','')  
   precio = precio.replace(',','')  
   preciodolar=anuncio.find('div',class_='pricedollars').text
   preciodolar = preciodolar.replace('$','')  
   preciodolar = preciodolar.replace(',','')
   preciodolar = preciodolar.replace('(','')  
   preciodolar = preciodolar.replace(')*','')
  
   tags = anuncio.find_all('td')           #Caracteristicas
   get_valores(tags)
   #print(nombre)
   #print(datos)
   
   registro = pd.DataFrame([datos])
   df = pd.concat([df,registro])
   
```

```python
df.to_csv(".\crautos.csv",index=False)
```
